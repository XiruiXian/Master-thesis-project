```python
import pandas as pd
```

```python
# patients to be focused on
Patient = ['chb04','chb06','chb08','chb15','chb17','chb19']
```

```python
for i in range(len(Patient)):
    df = pd.DataFrame(columns=['file name', 'number of seizures', 'seizure time/seconds'])
    file_name = []
    num_seizures = []
    start_seizures = []
    end_seizures = []
    seizure_time = []
    sub_seizure_time = []
    with open('/Volumes/Macintosh HD - Data/Master Thesis/chb-mit-scalp-eeg-database-1.0.0/'+Patient[i]+'/'+ Patient[i]+'-summary.txt','r') as f:
        lines = f.readlines()
        for line in lines:
            l       = line.strip()
            words   = l.split(':')
            if words[0] == 'File Name':
                file_name.append(words[1].strip().split('.')[0])
            elif words[0] == 'Number of Seizures in File':
                num_seizures.append(int(words[1]))
                if int(words[1]) == 0:
                    seizure_time.append(None)
            elif words[0].split(' ')[0] == 'Seizure':
                if words[0].split(' ')[-2] == 'Start':
                    start_seizures.append(int(words[-1].split(' ')[-2]))
                if words[0].split(' ')[-2] == 'End':
                    end_seizures.append(int(words[-1].split(' ')[-2]))
                    sub_seizure_time.append((start_seizures[-1],end_seizures[-1]))
                if len(sub_seizure_time) == num_seizures[-1]:
                    seizure_time.append(sub_seizure_time)
                    sub_seizure_time = []
    df['file name'] = file_name
    df['number of seizures'] = num_seizures
    df['seizure time'] = seizure_time
    display(df)
```
